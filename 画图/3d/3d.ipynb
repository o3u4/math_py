# 1. 点线 2d -> 3d

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
n = 10
xs = np.linspace(0, 100, n)
ys = np.linspace(100, 200, n)
fig, ax = plt.subplots()    # fig 表示整个图像, ax 为其中一个子图
ax.scatter(xs, ys, color="r")
ax.plot(xs, ys)
plt.show()
```

```python
zs = xs + ys  #增加一个维度，值为x+y的和
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.scatter(xs, ys, zs, color='r')
ax.plot(xs, ys, zs)
plt.show()
```

# 2. 面-plot_surface

```python
xs = np.arange(-10, 10, 0.5)
ys = np.arange(-10, 10, 0.5)
xs, ys = np.meshgrid(xs, ys)  #先生成网格坐标
zs = xs * (ys**3) - ys * (xs**3)  #计算网格中每个点的Z轴坐标
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_surface(xs, ys, zs)
plt.show()
```

# 3. 只显示网格-plot_wireframe

```python
xs = np.arange(-10, 10, 0.5)
ys = np.arange(-10, 10, 0.5)
xs, ys = np.meshgrid(xs, ys)  #先生成网格坐标
zs = xs * (ys**3) - ys * (xs**3)  #计算网格中每个点的Z轴坐标
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_wireframe(xs, ys, zs)
plt.show()
```

# 3. 立方体-voxels

```python
x, y, z = np.indices((10, 10, 8))   # 三组索引数组, 形状是 (10, 10, 8), 对应于 x、y、z 三个方向
cube1 = (x < 9) & (y < 9) & (z == 1)
cube2 = (x > 0) & (x < 8) & (y > 0) & (y < 8) & (z == 2)
cube3 = (x > 1) & (x < 7) & (y > 1) & (y < 7) & (z == 3)
cube4 = (x > 2) & (x < 6) & (y > 2) & (y < 6) & (z == 4)
cube5 = (x > 3) & (x < 5) & (y > 3) & (y < 5) & (z == 5)
cube = cube1 | cube2 | cube3 | cube4 | cube5
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
fig.set_size_inches(8, 6)       # 图表大小设置
ax.voxels(cube, color="goldenrod", edgecolor="g")
plt.show()
```

# 4. 更复杂的3d图

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
# mplot3d 为3d引擎, 适合复杂的3d图

# 创建一个空的Figure对象
fig = plt.figure()

# 不自动将创建的Axes3D对象添加到Figure中
ax = Axes3D(fig, auto_add_to_figure=False)

# 将创建的Axes3D对象手动添加到Figure中
fig.add_axes(ax)    # 只能添加一个图, 即为显示的图

# 生成数据
x = np.random.rand(100)
y = np.random.rand(100)
z = np.random.rand(100)

# 绘制散点图
ax.scatter(x, y, z)

plt.show()
```

# 5. 添加子图

```python
fig = plt.figure(figsize=(10, 5))

ax = fig.add_subplot(211, projection='3d')      # 211, 2行1列, 第1个子图  写成2, 1, 1也可
ax2 = fig.add_subplot(212, projection='3d')     # 211, 2行1列, 第2个子图

# 三维散点图
x = np.random.rand(50)  
y = np.random.rand(50)  
z = np.random.rand(50) 
ax.scatter(x, y, z, color='red', s=100)     # s=100, 散点大小

x2 = np.random.rand(100)
y2 = np.random.rand(100)
z2 = np.random.rand(100)
ax2.scatter(x2, y2, z2, color='blue', s=50)

# 保存图像
fig.savefig('6-2.png')
```
